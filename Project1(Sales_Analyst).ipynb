4. What Products are most often sold together?

```python
all_sales.head()
```

```python
# Perhaps we can first check if there are duplicate values inside the Order ID column
dup_order_id = all_sales['Order ID'].duplicated(keep=False)
```

```python
# Checking the index of the duplicated values
duplicate_rows_order_id = all_sales[dup_order_id]
```

```python
duplicate_rows_order_id
```

```python
duplicate_rows_order_id['Products'] = duplicate_rows_order_id.groupby('Order ID')['Product'].transform(lambda x : ','.join(x))
```

```python
duplicate_rows_order_id
```

```python
duplicate_rows_order_id = duplicate_rows_order_id[['Order ID', 'Products']].drop_duplicates()
```

```python
duplicate_rows_order_id
```

```python
count = Counter()

for row in duplicate_rows_order_id['Products']:
  row_list = row.split(',')
  count.update(Counter(combinations(row_list, 2)))
```

```python
# Better. But use .most_common() to see the most common occurence value
count
```

```python
# iPhone and Lightning Charging Cable is the most bought pair
for key, value in count.most_common(10):
  print(key, value)
```
